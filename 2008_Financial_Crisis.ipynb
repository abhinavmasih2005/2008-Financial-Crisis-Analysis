## 1. Importing Libraries

```python
# Ranks the entries by GDP level, with 1 being the highest value
df['GDP_Rank'] = df['GDP'].rank(ascending=False)

# To sort and view top 5
df.sort_values(by='GDP_Rank').head()
```

```python
# Categorize rows into recession and comparative boom
def categorize_recession(val):
    if val == 1:
        return 'recession'
    elif val == 0:
        return 'boom'
    else:
        return 'invalid'

df['Recession_Level'] = df['Recession'].apply(categorize_recession)
print(df['Recession_Level'].value_counts())
```

## 4. Assigning Dates to the Dataset

```python
# Generate a monthly date range starting from April 1990
date_range = pd.date_range(start='1990-04-01', periods=344, freq='ME')

# Add the date range as a new column
df['Date'] = date_range
df.head()
```

## 5. Data Visualization

### 5.1 Housing Index Over Time

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Housing_Index', data=df)
plt.title('Housing Index Over Time')
plt.xlabel('Date')
plt.ylabel('Housing Index')
plt.axvspan(pd.to_datetime('2007-01'), pd.to_datetime('2009-12'), color='green', alpha=0.2, label='Crisis Period')
plt.legend()
plt.show()
```

### 5.2 Federal Interest Rate Over Time

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Rate', data=df)
plt.title('Federal Interest Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Federal Interest Rate')
plt.axvspan(pd.to_datetime('2007-01'), pd.to_datetime('2009-12'), color='red', alpha=0.2, label='Crisis Period')
plt.legend()
plt.show()
```

### 5.3 GDP Over Time

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='GDP', data=df)
plt.title('GDP Over Time')
plt.xlabel('Date')
plt.ylabel('GDP')
plt.axvspan(pd.to_datetime('2007-01'), pd.to_datetime('2009-12'), color='yellow', alpha=0.2, label='Crisis Period')
plt.legend()
plt.show()
```

### 5.4 Industrial Production Over Time

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='INDPRO', data=df)
plt.title('Industrial Production Over Time')
plt.xlabel('Date')
plt.ylabel('Industrial Production')
plt.axvspan(pd.to_datetime('2007-01'), pd.to_datetime('2009-12'), color='violet', alpha=0.2, label='Crisis Period')
plt.legend()
plt.show()
```

### 5.7 Scatterplot: CPI vs GDP

```python
plt.figure(figsize=(14, 10))
sns.scatterplot(data=df, x='CPI', y='GDP')
plt.title('CPI vs GDP')
plt.grid(True)
plt.show()
```

### 5.8 Phase Division & Boxplot: S&P 500 Distribution Across Phases

```python
df['Phase'] = df['Date'].apply(lambda d: 'Pre-Crisis' if d < pd.to_datetime('2007-01') else
                                          'During Crisis' if d <= pd.to_datetime('2009-12') else 'Post-Crisis')

sns.boxplot(data=df, x='Phase', y='Price_x')
plt.title('S&P 500 Distribution Across Phases')
plt.show()
```

### 5.9 Boxplot: Housing Index Distribution Across Phases

```python
sns.boxplot(data=df, x='Phase', y='Housing_Index')
plt.title('Housing Index Distribution Across Phases')
plt.show()
```

### 5.10 Correlation Heatmap of Economic Indicators

```python
# Identifying treasury yield columns
t_bill_cols = ['3 Mo','4 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

# Creating a new column for average treasury yield
df['Avg_Treasury_Yield'] = df[t_bill_cols].mean(axis=1)

# Dropping the original treasury columns and the date column
df_heatmap = df.drop(columns=t_bill_cols+['Date'])

# Select only numeric columns
df_heatmap_numeric = df_heatmap.select_dtypes(include='number')

# Generating the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_heatmap_numeric.corr(), annot=True, cmap='coolwarm', annot_kws={"size": 12}, linewidths=0.5, linecolor='white')
plt.title('Correlation Between Economic Indicators (Including Avg T-Bill)', fontsize=16)
plt.tight_layout(pad=3.0)
plt.show()
```

### 5.11 Comparative Analysis of BBK Index Across Economic Phases

```python
# Make sure Date column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Segment into three phases
pre_crisis    = df[df['Date'] < '2007-01-01']
during_crisis = df[(df['Date'] >= '2007-01-01') & (df['Date'] <= '2009-12-31')]
post_crisis   = df[df['Date'] > '2009-12-31']

pre_mean    = pre_crisis['BBK_Index'].mean()
during_mean = during_crisis['BBK_Index'].mean()
post_mean   = post_crisis['BBK_Index'].mean()

plt.figure(figsize=(8,5))
plt.bar(['Pre-Crisis', 'During Crisis', 'Post-Crisis'],
        [pre_mean, during_mean, post_mean],
        color=['green', 'red', 'blue'])
plt.title('Average BBK_Index Across Phases')
plt.ylabel('BBK_Index')
plt.show()
```
